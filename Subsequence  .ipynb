```python
                    # With Numpy
import numpy as np
from datetime import datetime
startTime = datetime.now()
  # Convert spisok
spisok111=np.genfromtxt("10m.txt", delimiter='\t', dtype=np.float) 

   # First 4 tasks
max_value = np.max(spisok111)
min_value = np.min(spisok111)
median_value = np.median(spisok111)
average_value = np.average(spisok111)
 
    #Print 4 tasks
print('1.The maximum value of the array is: ', max_value)
print('2.The minimum value of the array is: ', min_value)
print('3.The median value of the array  is: ', median_value)
print('4.The average value of the array is: ', average_value)

  # Prog 1 -  Cycle UP
z_count=1
a=0
b=a+1
z_max=1
for element in spisok111:
    if b<(len(spisok111)) and  spisok111[a]<spisok111[b]:
        z_count+=1
        a+=1
        b+=1
    else:
        if z_count>z_max:
            z_max=z_count
        a+=1
        b+=1
        z_count=1

print('5.Max UP Subsequence               : ', z_max)

# Prog 2 -  Cycle DOWN
z_count=1
a=0
b=a+1
z_max=1
for element in spisok111:
    
    if b<(len(spisok111)) and  spisok111[a]>spisok111[b]:
        z_count+=1
        a+=1
        b+=1
    else:
        if z_count>z_max:
            z_max=z_count
        a+=1
        b+=1
        z_count=1
print('6.Max DOWN Subsequence             : ', z_max)
print('      Working time  is             : ', datetime.now() - startTime)
```
