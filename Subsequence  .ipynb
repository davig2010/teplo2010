```python
import numpy as np
from datetime import datetime
startTime = datetime.now()
  # Convert spisok
spisok=np.genfromtxt("10m.txt", delimiter='\t', dtype=np.float) 

        # First 4 tasks
    #1
max_value=None
min_value=None    
for element in spisok:
    current=element
    if max_value is None or max_value<current:
        max_value=current
print('1.The maximum value of the array is: ', max_value)
    #2
for element in spisok:
    current=element
    if min_value is None or min_value>current:
        min_value=current
print('2.The minimum value of the array is: ', min_value)
    #3
mid = len(spisok)//2
median_value = (spisok[mid]+spisok[mid-1])//2 if len(spisok)%2 ==0 else spisok[mid]
print('3.The median value of the array  is: ', median_value)
    #4
average_value=sum(spisok) / len(spisok)

print('4.The average value of the array is: ', average_value)

  # Prog 1 -  Cycle UP
z_count=1
a=0
b=a+1
z_max=1
for element in spisok:
    if b<(len(spisok)) and  spisok[a]<spisok[b]:
        z_count+=1
        a+=1
        b+=1
    else:
        if z_count>z_max:
            z_max=z_count
        a+=1
        b+=1
        z_count=1

print('5.Max UP Subsequence               : ', z_max)

# Prog 2 -  Cycle DOWN
z_count=1
a=0
b=a+1
z_max=1
for element in spisok:
    
    if b<(len(spisok)) and  spisok[a]>spisok[b]:
        z_count+=1
        a+=1
        b+=1
    else:
        if z_count>z_max:
            z_max=z_count
        a+=1
        b+=1
        z_count=1
print('6.Max DOWN Subsequence             : ', z_max)
print('      Working time  is             : ', datetime.now() - startTime)
```
